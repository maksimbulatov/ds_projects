Our idea seems to be right. Let us find the games with the same name on different platforms and fill the missing values. 

```python
def find_similar_names(data,column_name):
    """
    Function find the same names of the games with the missing values and different platforms.
    """
    full_values = data.loc[~data[column_name].isnull()].groupby('name')[column_name].count().sort_values(ascending=False).to_dict()
    missing_values = data.loc[data[column_name].isnull()].groupby('name')[column_name].count().sort_values(ascending=False).to_dict()
    names = []
    for element in full_values:
        if element in missing_values:
            names.append(element)
    return names

```

```python
names = find_similar_names(df,'year_of_release')
```

Create library with game names and year of release.

```python
names_dict = df.query('name in @names').groupby('name')['year_of_release'].max().to_dict()
names_dict
df['year_of_release'] = df['year_of_release'].fillna(df['name'].map(names_dict))
df.info()
```

By using this method we have managed to fill some missing values_ but we have still 146 games without year of release. Let us check these rows.

## 5. Check of hypothesis

- Average user ratings for Xbox One and PC platforms are the same;
- Average user ratings for Action and Sports genres are different.

Our null hypothesis is that the mean values are equal, and our alternative hypothesis is that they are not equal. We must use a t-test to test and see if the p-value is greater or less than alpha(0.05). If the p-value is greater than alpha, we allow for the null hypothesis. We can use the t-test on our data because the amount of data is large enough, even if the data does not have a normal distribution.

```python
names = ['XOne','PC']
games_fav=df.query('platform in @names')
games_fav = games_fav.dropna(subset=['user_score']).reset_index(drop=True)
H = games_fav.groupby('platform')['user_score']
H.describe()
```

```python
alpha = 0.05
results = st.ttest_ind(H.get_group('PC').values,
            H.get_group('XOne').values,
            equal_var = True)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print("Rejecting the null hypothesis")
else:
    print("It was not possible to reject the null hypothesis")
```

We reject the null hypothesis. We can accept the alternative hypothesis, where the average user rating is not the same for the XOne and PC platforms.

Let's do the same for the genres. The null hypothesis is that the average user rating is equal for the Action and Sports genres. The alternative is not equal.

```python
names = ['action','sports']
games_fav=df.query('genre in @names')
games_fav = games_fav.dropna(subset=['user_score']).reset_index(drop=True)
H = games_fav.groupby('genre')['user_score']
H.describe()
```

```python
alpha = 0.05
results = st.ttest_ind(H.get_group('action').values,
            H.get_group('sports').values,
            equal_var = True)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print("Rejecting the null hypothesis")
else:
    print("It was not possible to reject the null hypothesis")
```

Here we cannot reject the null hypothesis, the average user ratings are equal for the Action and Sports genres.
