### Model

Before we start training, we need to generate features for our model and partition our data into samples. But first let's predict the time series without training.

```python
train, test = train_test_split(df, shuffle=False, test_size=0.1)
pred_previous = test.shift().fillna(train['num_orders'].iloc[-1])
print("RMSE:", np.sqrt(mean_squared_error(test,pred_previous)))
```

We need a value no higher than 48, which would be much better, that is, if we achieve this metric, our model will be adequate.

Let's create a function that will create additional features for us, and we can also adjust the number of them.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['hour'] = data.index.hour
    #data['month'] = data.index.month
    #data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
data = df
make_features(data, 24, 2)
```

```python
train, test = train_test_split(df, shuffle=False, test_size=0.2)
valid, test = train_test_split(test, shuffle=False, test_size=0.5)

print(train.index.min(), train.index.max())
print(valid.index.min(), valid.index.max())
print(test.index.min(), test.index.max())
```

```python
train = train.dropna()
```

```python
train_features = train.drop(['num_orders'],axis=1)
train_target = train['num_orders']
valid_features = valid.drop(['num_orders'],axis=1)
valid_target = valid['num_orders']
```

#### Linear regression

```python
model_lin = LinearRegression()
model_lin.fit(train_features,train_target)
predicted_lin = model_lin.predict(valid_features)

print("RMSE:", np.sqrt(mean_squared_error(valid_target,predicted_lin)))
```

#### XGBoost

```python
model = XGBRegressor(objective='reg:squarederror', n_estimators=1000)
model.fit(train_features,train_target)
predicted = model.predict(valid_features)

print("RMSE:", np.sqrt(mean_squared_error(valid_target,predicted)))
```

```python
for max_depth in range(5,8,1):
    x = [.03, 0.05, .07]
    for learning_rate in x:
        model = XGBRegressor(objective='reg:squarederror', n_estimators=1000,max_depth = max_depth, learning_rate=learning_rate)
        model.fit(train_features,train_target)
        predicted = model.predict(valid_features)
        print('max_depth =', max_depth,'learning_rate =', learning_rate, ':', np.sqrt(mean_squared_error(valid_target,predicted)))
print('')
```

### Conclusion

We considered two models: linear regression and XGBoost. They showed good results on the validation sample. And XGBoost is better than the linear model. Let's see how they perform in the test sample.

### Testing

```python
test_features = test.drop(['num_orders'],axis=1)
test_target = test['num_orders']
```

#### Linear regression

```python
plt.figure(figsize=(15,7))
plt.plot(predicted_lin,'r',label = 'prediction')
plt.plot(test_target.values,'b',label = 'actual')
plt.legend()
plt.title('Prediction vs Real, RMSE:{:.2f}'.format(np.sqrt(mean_squared_error(test_target,predicted_lin))))
```
